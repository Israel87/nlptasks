```python
# application to understand some of the usages of NLP toolkit library
# import libraries 
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
# nltk.download()
from nltk import sent_tokenize
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Open text file
text_file = open('nlptexts.txt')

# Read the data 
text = text_file.read()

# Datatype of the data read :
print (type(text))
print ("\n")

# Print the text :
print(text)
print("\n")

# Length of the text :
print(len(text))
```

```python
# Tokenize the text by sentences :
sentences = sent_tokenize(text)

# How many sentences are there? :
print (len(sentences))

# Print the sentences :
# print(sentences)
sentences
```

```python
# Tokenize the texts with words :
words = word_tokenize(text)

# How many words are there :
print (len(words))

# Print words
print(words)
```

```python
# Find the frequency :
fdist = FreqDist(words)

# Print 10 most common words :
fdist.most_common(10)
```

```python
# visualize word distribution in our text
fdist.plot(10)
```

```python
# removing punctuation marks 
# Empty array to store words
words_no_punc = []

# removing punctuation marks :
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
        
# Print the words without punctuation marks :
print (words_no_punc)
print("\n")

# Length
print(len(words_no_punc))
```

```python
# frequency distribution using words without punctuation
fdist = FreqDist(words_no_punc)
fdist.most_common(10)
```

```python
# Plot the most common words on graph:
fdist.plot(10)
```

```python
# removing stopwords or coordinating conjunctions
# List of stopwords 
stopwords = stopwords.words("english")
print(stopwords)
print("\n")
print (len(stopwords))
```

```python
# Empty list to store clean words :
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
print(clean_words)
print("\n")
print(len(clean_words))
```

```python
fdist = FreqDist(clean_words)
fdist.most_common(10)
```

```python
# plot the graph of the clean words 
fdist.plot(10)
```

```python
# lemmatization, generating the dictionary word, 
# slower but generates words that makes sense.
lemmatizer = WordNetLemmatizer()

print (lemmatizer.lemmatize('studies'))
```

```python
# Stemming, truncating the word and quite faster than lemmatization 
# but generates word that may not make sense
stemmer = PorterStemmer()
print(stemmer.stem('studies'))
```

```python
# PoS tagging
tag = nltk.pos_tag(["Studying","Study", "Travel"])

print(tag)
```

```python
# PoS tagging example 
sentence = "A very beautiful young lady is walking on the beach"

# Tokenizing words :
tokenized_words = word_tokenize(sentence)

for words in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

tagged_words
```

```python
# Chunking example
# Extracting Noun Phrase from text :
# ? - optional character
# * - 0 or more repetitions

grammar = "NP : {<DT>?<JJ>*<NN>}"

# Creating a parser :
parser = nltk.RegexpParser(grammar)

# Parsing text :
output = parser.parse(tagged_words)
print(output)

# To visualize
output.draw()
```

```python
# Chinking example
# * - 0 or more repetitions 
# + - 1 or more repetitions 

# Here we are taking the whole string
# and then excluding adjectives from that chunk

grammar = r""" NP: {<.*>+} 
               }<JJ>+{"""

# Creating parser 
parser = nltk.RegexpParser(grammar)

# Parsing string :
output = parser.parse(tagged_words)
print(output)

# To visualize
output.draw()
```

```python
# Name Entity Recognition (NER):
# can automatically scan entire articles and pull out some fundamental entities 
# like people, organizations, places, date, time, money, and GPE discussed in them.
# Use cases : Content classification for news channels, Summarizing resumes, Optimizing search engine algorithms,
# Recommendation systems, Customer support

# Sentence for NER
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

# Tokenizing words :
tokenized_words = word_tokenize(sentence)

#PoS tagging :
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

# print (tagged_words)

# Named Entity Recognition :
# using binary = False shows the actual Name Entity.
N_E_R = nltk.ne_chunk(tagged_words, binary=False)

print(N_E_R)

# To visualize :
N_E_R.draw()
```

```python
# Bag of Words 

# Text for analysis 
sentences = ["Jim and Pam travelled by the bus.", "The train was late",
            "The flight was full. Travelling by flight is expensive"]

# Create an object :
cv = CountVectorizer()

# Generating output for Bag of Words :
B_O_W = cv.fit_transform(sentences).toarray()

# Total words with their index in model :
print(cv.vocabulary_)
print("\n")

# Features :
print(cv.get_feature_names())
print("\n")

# Show the output :
print(B_O_W)
```

```python
# TF-IDF
# Sentences for analysis :
sentences = ['This is the first document', 'This document is the second document']

# Create an object :
vectorizer = TfidfVectorizer(norm = None)

# Generating output for the TF_IDF :
X = vectorizer.fit_transform(sentences).toarray()

# Total words with their index in model :
print(vectorizer.vocabulary_)
print("\n")

# Features :
print(vectorizer.get_feature_names())
print("\n")

# Show the output :
print(X)
```
